```python
for field in shotlist:
    print("\\multicolumn{{5}}{{l}}{{{field}}}\\\\".format(field=field))
    
    for shot in shotlist[field]:
        if not shot.startswith("#"):        
            ii = t["datevobs"] == shot
            ra = t[ ii ]["ra"][0]
            dec = t[ ii ]["dec"][0]
            pa = t[ ii ]["pa"][0]
            n_ifu = t[ ii ]["n_ifu"][0]
            exptime = np.mean( t[ ii ]["exptime"][0] )

            c = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')


            s = "{} & {:.6f} & {:.6f} & {:.1f} & {:02d} & {:d}\\\\".format( shot, ra, dec , pa, n_ifu, int(round(exptime)))

            print(s)
```
