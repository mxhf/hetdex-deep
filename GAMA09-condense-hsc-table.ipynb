# Condenses HSC table down to (for us) relevant columns

Data obtained from
https://hsc-release.mtk.nao.ac.jp/datasearch/catalog_jobs
by
SELECT * FROM pdr2_wide.forced LEFT JOIN pdr2_wide.forced2 USING (object_id) 
LEFT JOIN pdr2_wide.photoz_mizuki USING (object_id) WHERE coneSearch(coord, 130.076800, 2.345986, 660)
