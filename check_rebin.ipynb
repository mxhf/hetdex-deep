# 01 - Check that all shots in a shotlist file have rebins for all their IFUs/amps

### Find all HETDEX shots, downselect by date and exposure time

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from astropy.io import ascii
from astropy.table import Table
import tables as tb
import os
```

```python
def ifuslot_amp_to_mffilename(s,a):
    fn="multi_xxx_{}_xxx_{}_rebin.pickle".format(s,a)
    return fn

def mffilename_to_ifuslot_amp(fn):
    #fn="multi_xxx_{}_xxx_{}_rebin.pickle".format(s,a)
    tt = fn.split("_")
    return tt[2],tt[4]
```

```python
fnshotlist = "/work/04287/mxhf/maverick/hetdex/cubes/shotlist_PCA_COSMOSBCD.txt"
rebindir = "/scratch/04287/mxhf/rebin2"
hdf5dir = "/work/03946/hetdex/hdr1/reduction/data/"
```

```python
shotlist = Table.read(fnshotlist, format='ascii.no_header')
```

```python
str( shotlist['col1'][0] )
```

```python
N = len(shotlist['col1'])

missing_ifus = []

for i, shot in enumerate(shotlist['col1']):
        filename = os.path.join( hdf5dir,  '{}.h5'.format(shot) )
        print("Processing shot {}/{}: {}".format(i,N,filename))
        fileh = tb.open_file(filename)
        t = Table(fileh.root.Data.Fibers[:])


        ff = []
        for s,a in zip( list( t["ifuslot"] ), list(t["amp"]) ):
            fn=ifuslot_amp_to_mffilename(s,a)
            for exp in [1,2,3]:
                ff.append(os.path.join(shot, "exp{:02d}".format(exp), fn))
        ff = np.unique(ff)

        for f in ff:
            path = os.path.join(rebindir, f)
            if not os.path.exists(path):

                ifuslot, amp = mffilename_to_ifuslot_amp(f)
                print("Did not find multifits for ifu {} amp {} expected: {}".format(ifuslot, amp, path) )

                missing_ifus.append(ifuslot)
```

```python
with open(fnshotlist + '.missing', 'w') as f:
    for ifuslot in np.unique(missing_ifus):
        f.write(ifuslot + '\n')
```
