```python
from astropy.io import fits
from astropy import wcs 
import numpy as np
from astropy.table import Table
import glob
```

```python
t = Table(dtype=[int,float,float], names=['ifuslot','ra','dec'])

clist=!ls data/outcube_median_???.fits.gz

clist = glob.glob("data/outcube_COSMOSD_???.fits.gz")
for f in clist:
    #print(f)
    ifu = f[21:24]
    # print(ifu)

    hdu = fits.open(f)

    w = wcs.WCS(hdu[0].header)

    __,ysize,xsize = hdu[0].data.shape

    #aastart,ddstart,_ = w.all_pix2world(0,0,0, 0)
    #aaend,ddend,_     = w.all_pix2world(0,xsize,ysize, 0)

    #aacenter,ddcenter,__ = w.all_pix2world(0,xsize/2.,ysize/2., 0)
    aacenter,ddcenter,__ = w.wcs_pix2world(xsize/2.,ysize/2., 0, 0)

    #ra_size  = (aaend - aastart)*np.cos( np.deg2rad (ddcenter) )
    #dec_size = (ddend - ddstart)

    ra,dec = float(aacenter),float(ddcenter)
    
    print(ifu, ra, dec)
    t.add_row([ifu, ra, dec])
```

```python
t.write("center_coor.txt", format="ascii.fast_tab", overwrite=True)
```
