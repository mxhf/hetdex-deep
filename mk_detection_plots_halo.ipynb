```python
from astropy.io import ascii
import numpy as np
```

```python
from IPython.parallel import Client
c = Client()
lv = c.load_balanced_view()   # this object represents the engines (workers)
```

```python
field = "COSMOSD"
c = ascii.read("data/mmsf2outcube_COSMOSD_allifu.cat")
ii = c["class2"] == "halo"
cat = c[ii]
cat.sort(keys=["ifu","id"])
print("Generating images for {} objects.".format(np.sum(ii)))
```

```python
def mk_detection_plots(field, ifu, objid, DRYRUN=False):
    import os
    cmd="python src/mk_detection_plots2.py -f {field} -i {ifu} -j {objid} -p _halo --catalog data/mmsf2outcube_{field}_allifu.cat > foo.log"\
        .format(field=field, ifu=ifu, objid=objid)
    if DRYRUN:
        return cmd
    
    return os.system( cmd )


tasks = []
for ifu, objid in zip(cat['ifu'], cat['id']):
    tasks.append( lv.apply(mk_detection_plots, field, ifu, objid) )
    
result = [task.get() for task in tasks]  # blocks until all results are back

for r in result:
    print(r)
```
