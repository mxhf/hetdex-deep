```python
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
import numpy as np
```

```python
N = 1000

R = 20  # readnoise
S = 10 # signal
K = 10 # sky
rn = np.random.normal(size = N, scale = R)
sn = np.random.poisson(S, size = N)
kn = np.random.poisson(K, size = N)
s = rn + sn + kn
```

```python
plt.plot(s, drawstyle="steps-mid")
print("noise: ", np.std(s))
print("SN: ", float(S)/np.std(s))
```

```python
nn = 10.**( np.linspace(np.log10(1),np.log10(100),5 ) )

SNSN = []
for n in nn:
    RN = []
    SN = []
    KN = []
    for i in range(int(n)):
        rn = np.random.normal(size = N, scale = R)
        sn = np.random.poisson(S, size = N)
        kn = np.random.poisson(K, size = N)
        RN.append(rn)
        SN.append(sn)
        KN.append(kn)
    RN = np.array(RN)
    SN = np.array(SN)
    KN = np.array(KN)
    
    s = np.mean(RN + SN + KN, axis=0)
    print("noise: ", np.std(s))
    print("SN: ", float(S)/np.std(s))
    SNSN.append(float(S)/np.std(s))
    
plt.plot(nn, SNSN)
plt.xlabel("")
```
