```python
from astropy.io import ascii
import numpy as np

c = ascii.read("../data/mmsf2outcube_COSMOSC_allifu.cat")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

params = {'legend.fontsize': '16',
          'figure.figsize': (15, 5),
         'axes.labelsize': '28',
         'axes.titlesize':'28',
         'xtick.labelsize':'28',
         'ytick.labelsize':'28'}

plt.rcParams.update(params)
```

```python
isplit_good = c["split_good"] == "True"

ipointsource = c["class2"] == "pointsource"

ishowcase = c["showcase"] 

ii = ipointsource * isplit_good
```

```python
plt.hist(c["flux"][ii]*1e17, bins=np.arange(0,15,1))
plt.xlabel("flux [erg / (cm^2 s)]")
plt.ylabel("N")
#plt.xscale("log")
#plt.yscale("log")
```

```python
f = plt.figure(figsize=[5,5])
plt.plot(c["ra_com"][ii],c["dec_com"][ii],'x')
plt.axis("equal")
plt.xlabel("RA [deg]")
plt.ylabel("Dec [deg]")
```

```python
from astropy.table import Table
```

```python
t = Table(c[ii])
```

```python
t.write("cosmosc_pointsources.ecsv", format="ascii.ecsv")
```
